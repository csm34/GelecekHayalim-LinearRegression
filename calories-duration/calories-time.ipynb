```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('last_version.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Duration'], df['Calories'])
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Calories vs Duration')
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(df[['Duration']], df['Calories'])
```

```python
predicted = model.predict(df[['Duration']])
```

```python
predicted
```

```python
plt.scatter(df['Duration'], df['Calories'], color='blue', label='Actual Data')
plt.plot(df['Duration'], predicted, color='red', label='Linear Regression')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Calories vs Duration with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# parameters
model.coef_, model.intercept_
```

```python
# Error Calculation
mae = mean_absolute_error(df['Calories'], predicted)
mse = mean_squared_error(df['Calories'], predicted)
rmse = np.sqrt(mse)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Calculate the Mean Absolute Error
mae = mean_absolute_error(df['Duration'], predicted)

# Calculate the R-squared score
r2 = r2_score(df['Duration'], predicted)

mae, r2
```

```python
# Print the error metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
