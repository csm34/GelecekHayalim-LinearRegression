```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('last_version.csv')
```

```python
np.random.seed(0)
random_sample = df.sample(n=50)
```

```python
random_sample.to_csv('last_version.csv', index=False)
```

```python
# Scatter plot of Calories vs Duration
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration'], df['Calories'], alpha=0.7)
plt.title('Calories vs Duration')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(df[['Calories']], df['Duration'])
```

```python
predicted = model.predict(df[['Calories']])
predicted
```

```python
# Scatter plot of actual data
plt.figure(figsize=(10, 6))
plt.scatter(df['Calories'], df['Duration'], alpha=0.7, label='Actual Data')

# Line plot of predicted data
plt.plot(df['Calories'], predicted, color='red', label='Linear Regression Model')

# Adding title and labels
plt.title('Calories vs Duration with Linear Regression Model')
plt.xlabel('Calories')
plt.ylabel('Duration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# modelimizin parametrelerini görelim
model.coef_, model.intercept_
```

```python
from sklearn.metrics import mean_squared_error

# Calculate the Mean Squared Error
mse = mean_squared_error(df['Duration'], predicted)
mse
```
